```python
import yfinance as yf
import pandas as pd
```

```python
import os
proxy = 'http://127.0.0.1:7890' # 代理设置，此处修改
os.environ['HTTP_PROXY'] = proxy
os.environ['HTTPS_PROXY'] = proxy
```

```python
# 彻底禁掉代理（见上一节）
for v in ("HTTP_PROXY","HTTPS_PROXY","http_proxy","https_proxy"):
    os.environ.pop(v, None)
```

```python
def get_all_extracted_stocks(ticker_list, start_date, end_date):
    """
    批量获取并返回所有股票的每日收盘价
    """
    print(f"正在下载 {len(ticker_list)} 只股票的数据...")
    
    # 批量下载 (Threads=True 开启多线程)
    df = yf.download(ticker_list, start=start_date, end=end_date, auto_adjust=True, threads=True)
    
    # 提取收盘价列
    if isinstance(df.columns, pd.MultiIndex):
        closing_prices = df['Close']
    else:
        closing_prices = df[['Close']]
        
    return closing_prices
```

```python
# --- 准备合并后的代码列表 ---
# 自动去重并剔除可能包含非法字符的代码
all_tickers = list(set([
    "PLTR", "ONDS", "AVAV", "KTOS", "DPRO", "UMAC", "RCAT", "NOC", "RTX", "LMT",
    "LSCC", "MBLY", "MRVL", "VRT", "AMBA", "HPS", "AAOI", "ACHR", "AEVA", "AG", 
    "ALBT", "AMZN", "AMD", "AREC", "ARQQ", "ASML", "ASNS", "ASTS", "BABA", "BBAI", 
    "BE", "BEP", "BIDU", "BMNR", "DMRN", "BTDR", "BYND", "CACI", "CHTR", "CIFR", 
    "CLSK", "CRML", "CVX", "DJT", "ELVA", "EVGO", "EVLV", "FFIE", "FIX", "FLY", 
    "GE", "GGLL", "GOOGL", "GPUS", "HIMX", "HOOD", "HUT", "HYMC", "INTC", "IREN", 
    "JOBY", "JMIA", "KRBNF", "LEGN", "LLY", "LUNR", "MARA", "MSTR", "MU", "NDRA", 
    "NET", "NFLX", "NIU", "NVO", "NVDA", "NXU", "OKLO", "OPEN", "OSCR", "OSS", 
    "PATH", "PLG", "POET", "PPPX", "PRME", "QQQ", "QS", "RDDT", "RGTI", "RKLB", 
    "ROIV", "SCHL", "SE", "SIDU", "SKYT", "SMR", "SNPS", "SOFI", "SPCE", "SPGI", 
    "SPIR", "SQQQ", "SUEK", "SYM", "TBLT", "TDOC", "TSLA", "TSM", "U", "UNV", 
    "UUUU", "V", "VNM", "WOLF", "ZETA"
]))

# 设置时间范围
start = "2024-01-01"
end = "2026-01-15"
```

```python
# 获取数据
all_prices = get_all_extracted_stocks(all_tickers, start, end)

# 打印结果预览（前5行）
print("\n--- 批量收盘价数据预览 ---")
print(all_prices.head())

# 保存为 CSV 文件方便后续分析
```
